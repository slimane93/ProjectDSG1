# World Happiness Report

## Introduction 
Le bonheur est un état de satisfaction qui se caractérise par sa stabilité et sa durée. Le World Happiness Report est une mesure du bonheur réalisée par le United Nations Sustainable Development Solutions Network chaque année. Ces rapports sont publiés chaque année depuis 2012 et classe 155 apys selon le niveau de bonheur. On peut y retrouver plusieurs axes d'analyses tels que : 
 - Les Pays 
 - Les régions 
 - Le rang du bonheur de chaque pays 
 - La moyenne du bonheur des citoyens (moyenne entre 0 et 10)
 - La liberté 
 - La confiance 
 - Espérance de vie
 - ...
 
## Problématique 
Au cours de ce projet, nous analyserons cette problématique :

<div class="alert-danger">
Quels sont les pays ou régions qui ont amélioré le plus leur état d’esprit de 2015 à 2019 ?
</div>

Pour répondre à cette problématique, plusieurs questions viennent à nous !

 - Quelles sont les pays qui ont pu faire évoluer leur taux de bonheur ? et inversement. 
 
 - Quel est le facteur le plus faible ? ( de manière générale )
 
 - Quelle est la pire année pour chaque continent ? 
 
 - ... D'autres questions qu'on verra durant l'analyse.

https://www.kaggle.com/unsdsn/world-happiness

## Carte du bonheur en fonction des années

Les cartes permettent de nous donner un aperçu général du bonheur dans le monde. On se rend compte que le bonheur fluctue très peu d'une année à l'autre. Nous pouvons quand même en tirer que les régions du monde les plus heureuses sont les Amériques, l'Europe de l'ouest et l'Océanie.

```python
def afficher_carte(annee,titre):
    data = dict(
            type = 'choropleth',
            colorscale = 'Viridis',
            marker_line_width=1,
            locations = annee['Country'],
            locationmode = "country names",
            z = annee['Happiness Score'],
            text = annee['Country'],
            colorbar = {'title' : 'Happiness Score'})
    layout = dict(title = titre,
              geo = dict(projection = {'type':'equirectangular'}, showocean = False, showframe = True, showlakes = True, showrivers = True, )
             )
    choromap = go.Figure(data = [data],layout = layout)
    iplot(choromap,validate=False)
    
afficher_carte(df2015,'Happiness Map en 2015')
afficher_carte(df2016,'Happiness Map en 2016')
afficher_carte(df2017,'Happiness Map en 2017')
afficher_carte(df2018,'Happiness Map en 2018')
afficher_carte(df2019,'Happiness Map en 2019')
```

## Treemap en fonction des années

Tout comme les cartes, les Treemap permettent de visualiser le score de bonheur de chaque pays, regions, continent.

Les différents tableaux où sont stockés les données non pas la même liste de pays et certains non pas la région de renseigné, donc nous prenons la liste de 2015 car c'est celle qui contient le plus de pays et les régions sont bien renseignées.
Pour les pays qui ne correspondent pas, ils auront 'Other' comme région.

```python
def treemap(annee,titre): # Recuperer les regions selon 2015
    liste_region=[]
    compter=df2015['Country'].value_counts() # Création d'une liste contenant le nombre d'itération de chaque pays donc 1
    for i in range(len(annee)):
        try:
            # si la valeur d'un pays est égal à 0 donc elle n'est pas dans df2015['Country'] se qui entraine une erreur
            # car les listes n'auront pas les même dimensions 
            if compter[annee.loc[i,'Country']] > 0: 
                for j in range(len(df2015)):
                    if annee.loc[i,'Country']==df2015.loc[j,'Country']:
                        liste_region.append(df2015.loc[j,'Region']) 
        except: 
            liste_region.append('Other')
    

    df= pandas.DataFrame(
        dict(Region=liste_region,Country=annee['Country'],Happiness_Rank=annee['Happiness Rank'],Happiness_Score=annee['Happiness Score']))
    df["world"] = "world" # in order to have a single root node
    fig = px.treemap(df,title=titre ,path=['world','Region', 'Country'], values='Happiness_Score',
                      color=df['Happiness_Score'], hover_data=['Happiness_Rank'],
                      color_continuous_scale='RdBu',
                      color_continuous_midpoint=np.average(df['Happiness_Score'], weights=df['Happiness_Score']))
    
    fig.show()
    
treemap(df2015,'treemap de 2015')
treemap(df2016,'treemap de 2016')
treemap(df2017,'treemap de 2017')
treemap(df2018,'treemap de 2018')
treemap(df2019,'treemap de 2019')
```

## Facteur qui rend le plus heureux

Pour connaitre le facteur qui rend le plus heureux nous avons choisi de compter le nombre de fois que ce facteur est le plus influent (donc le plus grand en terme de score) par année.

Les fichiers contenants les données non pas exactement les même colonnes d'une année à l'autre, afin de pouvoir parcourir les différents tableaux il était nécessaire d'écrire dans le programme une condition qui nous permettait d'ignorer les colonnes qui ne nous intéressent pas.

On constate que se qui rend le plus heureux dans un pays est un haut PIB par habitant et la famille.

```python
def facteur_plus_important(tableau):
    maxi=0
    facteur=[]
    liste=[]
    # Je ne veux pas prendre en compte les valeurs des colonnes indiqués dans exeption 
    exeption=['Region','Country','Standard Error','Happiness Rank','Happiness Score','Upper Confidence Interval','Lower Confidence Interval','Dystopia Residual']
    for i in range(len(tableau)):
        pays=tableau.loc[i]["Country"]
        for j in range(0,tableau.shape[1]): #shape[1] permet de connaitre le nombre de colonnes
            if tableau.columns[j] not in exeption:
                if tableau.loc[i][j] >= maxi :
                    maxi = tableau.loc[i][j]
                    colonne = tableau.columns[j]
                liste=[pays,maxi,colonne]
        facteur.append(liste)
    return np.array(facteur)

majeur_2015=facteur_plus_important(df2015)
majeur_2016=facteur_plus_important(df2016)
majeur_2017=facteur_plus_important(df2017)
majeur_2018=facteur_plus_important(df2018)
majeur_2019=facteur_plus_important(df2019)




# histogramme des facteurs les plus impactants 

    # Je récupere la liste des facteurs
facteur_majeur_2015=majeur_2015[:,2]
facteur_majeur_2016=majeur_2016[:,2]
facteur_majeur_2017=majeur_2017[:,2]
facteur_majeur_2018=majeur_2018[:,2]
facteur_majeur_2019=majeur_2019[:,2]

    # Afficher les figures
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,figsize=(12, 20))
ax1.hist(facteur_majeur_2015)
ax1.set_title('Facteur qui rend le plus heureux en 2015')
ax2.hist(facteur_majeur_2016)
ax2.set_title('Facteur qui rend le plus heureux en 2016')
ax3.hist(facteur_majeur_2017)
ax3.set_title('Facteur qui rend le plus heureux en 2017')
ax4.hist(facteur_majeur_2018)
ax4.set_title('Facteur qui rend le plus heureux en 2018')
ax5.hist(facteur_majeur_2019)
ax5.set_title('Facteur qui rend le plus heureux en 2019')
plt.show()



```

## Facteur qui influe le moins

Nous avons appliqué le même raisonnement que pour déterminer le facteur qui rend le plus heureux.

En revanche les facteurs qui influent le moins sur le bonheur sont la confiance envers le gouvernement (corruption) et la générosité.

```python
def facteur_moins_important(tableau):
    liste=[]
    facteur=[]
    for i in range(len(tableau)):
        pays=tableau.loc[i]["Country"]
        # Je ne veux pas prendre en compte les valeurs des colonnes indiqués dans exeption 
        exeption=['Region','Country','Standard Error','Happiness Rank','Happiness Score','Upper Confidence Interval','Lower Confidence Interval','Dystopia Residual']
        mini=tableau.loc[i]["Happiness Score"] #Je prend la fourchette haute pour etre sur d'avoir la plus grande valeur du tableau
        for j in range(1,tableau.shape[1]): #shape[1] permet de connaitre le nombre de colonnes
            if tableau.columns[j] not in exeption:
                if tableau.loc[i][j] <= mini :
                    mini = tableau.loc[i][j]
                    colonne = tableau.columns[j]
                liste=[pays,mini,colonne]
        facteur.append(liste)
    return np.array(facteur)

mineur_2015=facteur_moins_important(df2015)
mineur_2016=facteur_moins_important(df2016)
mineur_2017=facteur_moins_important(df2017)
mineur_2018=facteur_moins_important(df2018)
mineur_2019=facteur_moins_important(df2019)

# histogramme des facteurs les plus impactants 

    # Je récupere la liste des facteurs
facteur_mineur_2015=mineur_2015[:,2]
facteur_mineur_2016=mineur_2016[:,2]
facteur_mineur_2017=mineur_2017[:,2]
facteur_mineur_2018=mineur_2018[:,2]
facteur_mineur_2019=mineur_2019[:,2]


    # Afficher les figures
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,figsize=(12, 20))
ax1.hist(facteur_mineur_2015)
ax1.set_title('Facteur qui rend le moins heureux en 2015')
ax2.hist(facteur_mineur_2016)
ax2.set_title('Facteur qui rend le moins heureux en 2016')
ax3.hist(facteur_mineur_2017)
ax3.set_title('Facteur qui rend le moins heureux en 2017')
ax4.hist(facteur_mineur_2018)
ax4.set_title('Facteur qui rend le moins heureux en 2018')
ax5.hist(facteur_mineur_2019)
ax5.set_title('Facteur qui rend le moins heureux en 2019')
plt.show()
```

## Impact de chaques facteurs sur le bonheur en fonction des années

Les diagrammes affichent l'impact de chaque facteur sur le bonheur, pour avoir cela nous avons fait la moyenne mondiale de chaque facteur.

Les facteurs qui influent à 75% sur le bonheur peu importe l'année sont l'économie, la famille et la santé et leurs part augmente d'année en année.

```python
# Diagramme de l'impact de chaque facteur (moyenne)

def moyenne_facteur(tableau):
    liste=[]
    label=['Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Generosity','Trust (Government Corruption)']
    for i in range(len(label)):
        val = tableau.loc[:,label[i]]
        somme = val.sum()
        ligne = tableau.shape[0]
        moy = somme/ligne
        liste.append(moy)
    return liste

moyenne_facteur_2015=moyenne_facteur(df2015)
moyenne_facteur_2016=moyenne_facteur(df2016)
moyenne_facteur_2017=moyenne_facteur(df2017)
moyenne_facteur_2018=moyenne_facteur(df2018)
moyenne_facteur_2019=moyenne_facteur(df2019)


labels = ['Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Generosity','Trust (Government Corruption)']

fig,ax = plt.subplots(2,2,figsize=(12, 10))
fig.suptitle('Impact des différents facteurs sur le bonheur',fontsize=20)
ax[0,0].pie(moyenne_facteur_2015, autopct = lambda x: str(round(x, 2)) + '%')
ax[0,0].set_title('En 2015')
ax[0,1].pie(moyenne_facteur_2016, autopct = lambda x: str(round(x, 2)) + '%')
ax[0,1].set_title('En 2016')
ax[1,0].pie(moyenne_facteur_2017, autopct = lambda x: str(round(x, 2)) + '%')
ax[1,0].set_title('En 2017')
ax[1,1].pie(moyenne_facteur_2018, autopct = lambda x: str(round(x, 2)) + '%')
ax[1,1].set_title('En 2018')
fig.legend(labels)
plt.show()

plt.pie(moyenne_facteur_2019, autopct = lambda x: str(round(x, 2)) + '%')
plt.title('En 2019')

plt.tight_layout()
plt.show()
```

## Régression linéaire

Nous prenons en compte que les relations avec un coefficient de corrélation supérieure à 0,7.

```python
def regression(annee,titre):   
    fig , axes = plt.subplots(2,3,figsize=(15,7))

    g1 = sns.regplot(x=annee['Family'],y=annee['Economy (GDP per Capita)'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[0,0])
    g1.set_ylabel('Economy (GDP per Capita)')
    g1.set_xlabel('Family')

    g2 = sns.regplot(y=annee['Happiness Score'],x=annee['Family'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[0,1])
    g2.set_xlabel('Family')
    g2.set_ylabel('Happiness Score')

    g3 = sns.regplot(y=annee['Happiness Score'],x=annee['Economy (GDP per Capita)'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[0,2])
    g3.set_xlabel('Economy (GDP per Capita)')
    g3.set_ylabel('Happiness Score')

    g4 = sns.regplot(y=annee['Happiness Score'],x=annee['Health (Life Expectancy)'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[1,0])
    g4.set_xlabel('Health (Life Expectancy)')
    g4.set_ylabel('Happiness Score')

    g5 = sns.regplot(x=annee['Family'],y=annee['Health (Life Expectancy)'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[1,1])
    g5.set_ylabel('Health (Life Expectancy)')
    g5.set_xlabel('Family')

    g6 = sns.regplot(x=annee['Economy (GDP per Capita)'],y=annee['Health (Life Expectancy)'],ci=None,scatter_kws = {'color':'r','s':9},ax=axes[1,2])
    g6.set_ylabel('Health (Life Expectancy)')
    g6.set_xlabel('Economy (GDP per Capita)')

    fig.suptitle(titre, fontsize=16)
    plt.show()

regression(df2015,'2015')
regression(df2016,'2016')
regression(df2017,'2017')
regression(df2018,'2018')
regression(df2019,'2019')
```
